# 추천 시스템 구현 및 최적화

이 책은 추천 시스템의 기본 개념부터 고급 최적화 기법까지 단계별로 구현하고 평가하는 과정을 담고 있습니다. MovieLens 데이터셋을 사용하여 다양한 추천 알고리즘을 구현하고, 이들의 성능을 비교 분석합니다.

## 목차

### [1. 소개 및 기초 개념](chapters/01_introduction.ipynb)
추천 시스템의 기본 개념, 목표, 그리고 이 책에서 다룰 내용에 대한 개요를 제공합니다.

### [2. 데이터 준비 및 전처리](chapters/02_data_preparation.ipynb)
MovieLens 데이터셋 로드 및 전처리 과정을 설명하고 훈련/테스트 데이터 분할 방법을 다룹니다.

### [3. 평가 메트릭](chapters/03_evaluation_metrics.ipynb)
RMSE, Precision@K, Recall@K 등 추천 시스템 성능 평가를 위한 지표들을 구현하고 해석합니다.

### [4. 베이스라인 추천 알고리즘](chapters/04_baseline_recommenders.ipynb)
전체 평균 기반, 인기도 기반 등 기본적인 추천 시스템을 구현하고 기준선을 설정합니다.

### [5. 협업 필터링](chapters/05_collaborative_filtering.ipynb)
사용자 기반(UserKNN)과 아이템 기반(ItemKNN) 협업 필터링 알고리즘을 구현하고 비교합니다.

### [6. 행렬 분해 기법](chapters/06_matrix_factorization.ipynb)
잠재 요인 모델링을 활용한 행렬 분해(Matrix Factorization) 추천 시스템을 구현합니다.

### [7. 알고리즘 비교 및 평가](chapters/07_evaluation_comparison.ipynb)
구현된 다양한 알고리즘의 성능을 체계적으로 비교하고 분석합니다.

### [8. 최적화 기법 및 고급 주제](chapters/08_optimization_techniques.ipynb)
희소 행렬, 하이브리드 모델, 효율적인 Top-K 추천 생성 등 고급 최적화 기법을 다룹니다.

## 성능 요구사항

구현된 알고리즘은 다음의 성능 기준을 충족해야 합니다:

- **필수 요건**: 
  - RMSE: Global Mean 대비 개선
  - Precision@10/Recall@10: Popularity 대비 개선

- **권장 목표**:
  - RMSE: Global Mean 대비 ≥5% 개선
  - Precision/Recall@10: Popularity 대비 ≥10% 개선
